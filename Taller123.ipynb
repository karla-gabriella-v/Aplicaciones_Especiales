# Variable Mass
## Theory
Systems that gain or lose mass: A system that gains mass by continuously absorbing particles or loses mass by continuously expelling particles.

$Fig. 1$
![title](masavariable.png)

To apply the principle of momentum and quantity of motion to the analysis of this system one must consider at time $t$ the system $S$ plus the particles of mass $\Delta m$ that absorbs $S$ during the time interval $\Delta t$. The velocity of $S$ at time $t$ is denoted by $v$, the velocity of $S$ at time $t+ \delta t$ is denoted by $v + \Delta v$ and the absolute velocity of the absorbed particles is denoted by $v_a$. Applying the principle is written as:

$$mv + (\Delta m) v_0 +\sum F \Delta t = \left ( m + \Delta m \right )+\left ( v + \Delta v \right )$$

Whence:

$$\sum {F} \Delta t= m\Delta v+\Delta m\left( v-v_{a} \right)+\left( \Delta m \right)\left( \Delta v \right)$$

If the relative velocity $u$ with respect to $S$ of the absorbing particles is introduced, it is written $a =v_a-v$, considering $v_a$ < $v$. If the last term in equation 2 which is of second order is ignored, it follows:

$$\sum F\Delta t = m\Delta v-\left ( \Delta m \right )u$$

Dividing by $\Delta t$ and letting $\Delta t$ tends to zero gives the limit:

$$\sum F = m\frac{dv}{dt}-\frac{dm}{dt}u$$

Regrouping the terms and recalling that $\frac{dv}{dt}=a$, where $a$ is the acceleration of the system $S$ one writes:

$$\sum F+\frac{dm}{dt}u = ma$$

Which shows that the action on $S$ of the particles being absorbed is equivalent to a thrust:

$$P=\frac{dm}{dt}u$$

Which tends to slow down the motion of $S$ since the relative velocity $u$ of the particles of is directed to the left. In the International System the thrust corresponds to Newtons.

The above is also used to determine the motion of a system $S$ that loses mass. In this case the rate of change of mass is negative and the action on $S$ of the particles being ejected is equivalent to a thrust in the direction of $-u$, i.e. in the direction opposite to that in which the particles are being ejected. A rocket represents a characteristic case of a system that continuously loses mass. 

*When the absolute velocity $v_a$ of the absorbing particles is zero, $u=-v$ and the formula (3) becomes

$$\sum{F}=\frac{d}{dt}\left( mv \right)$$

By comparing the formula obtained, it is observed that it is possible to apply Newton's second law to a system that gains mass as long as the absorbed particles are initially at rest. It can also be applied to systems that lose mass as long as the velocity of the ejected particles is zero with respect to the chosen reference frame.*

## Exercise
A rocket of initial mass $m_{0}$ (including shell and fuel) is launched vertically at time $t_{0}$ . Fuel is consumed at a constant rate $q = \frac{dx}{dt}$ and is ejected at a constant speed u relative to the rocket. Obtain an expression for the magnitude of the rocket's velocity at time $t$, ignoring air resistance.

Solution:
At time t, the mass of the rocket shell and the remaining fuel is: $m = m_0- q \cdot t$ and the velocity is $v$. During the time interval, a mass of fuel is expelled $\Delta m=q *\Delta t$ with a speed $u$ relative to the rocket. Denoted by $v_{e}$ the absolute velocity of the ejected fuel, the principle of impulse and the momentum between time $\mathcal{t}$ and time $t+\Delta t$.

$Fig. 2$
![title](dclg3.jpeg)

That is:

$$\left (m_0- q \cdot t  \right )v - g \left (m_0- q \cdot t  \right ) \Delta t = \left (m_0- q t -q\Delta t  \right )\left ( v+\Delta v \right )-q\Delta t\left ( u-v \right )$$

By dividing variables by $\Delta t$ and letting $\Delta t$ go to zero, we obtain:

$$-g (m_0-qt) = (m_0-qt)\frac{dv}{dt}-qu$$

By separating variables and integrating from $t = 0$, $v = 0$ to $t = t$, $v = v$,

$$dv=\left ( \frac{qu}{m_0-qt} -g \right ) dt$$

$$\int_{0}^{v}\left ( \frac{qu}{m_0-qt} -g \right ) dt$$ 

$$\mathcal{v}=\left[ -\mathcal{u}\cdot  ln\left( m_{o}-qt \right)-gt \right] \mathrm{|}_{0}^{t}$$

$$v = - u \cdot ln (\frac{m_0}{m_0-qt}) – gt$$ 

### *Commentary.*
The mass remaining in time $t_f$, after all the fuel has been consumed, is equal to the mass of the rocket shell:

$$m_s = m_0-q \cdot t_f$$

and the maximum speed that the rocket reaches is:

$$v_m = u \cdot ln \left ( \frac{m_0}{m_s} \right ) - g \cdot t_f$$

Assuming that the fuel is expelled in a relatively short period, the term $g \cdot t_f$ is small and we have:

$$v_m \approx  u \cdot ln \left ( \frac{m_0}{m_s} \right )$$

To escape the Earth's gravitational field, a rocket must reach a speed of $11.18 km / s$. Assuming that $u = 2200m / s$ and $v_m = 11.18 \frac{km}{s}$, we obtain  $\frac{m_s}{m_0} = 161$.

Consequently, to launch each kilogram of the rocket frame into space, it is necessary to consume more than $161kg$ of fuel, if an impeller is used that produces $u = 2200 m/s$.
